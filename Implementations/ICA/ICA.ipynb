# Independent Component Analysis

```python
#importing necessary libraries
import numpy as np
import wave
```

In this notebook we are going to work with audio file.

```python
#loading the audio file
audio1=wave.open("D:/Ai/Data_science/DSND_Term1-master/lessons/Unsupervised/5_ICA/ICA mix 1.wav",'r')
```

```python
#taking a look at the parameters
audio1.getparams()
```

```python
#Extracting the frame of the audio and converting it into integer format
signal_1_string=audio1.readframes(-1)
signal_1=np.fromstring(signal_1_string,'Int16')
```

```python
#finding the length and printing the first 50 integers of the frame
print('No. of frame:',len(signal_1))
print('Printing first 50 elements:',signal_1[:50])
```

```python
#To find the total time of the audio
264515/44100
```

```python
#We will use matplotlib to plot the whole audio
import matplotlib.pyplot as plt

fr=audio1.getframerate()
time=np.linspace(0,len(signal_1)/fr,len(signal_1))

plt.figure(figsize=(12,2))
plt.plot(time,signal_1,c="#3ABFE7")
plt.ylim(-35000, 35000)
plt.show()
```

```python
#similarly doing this for all of the audio file
audio2=wave.open("D:/Ai/Data_science/DSND_Term1-master/lessons/Unsupervised/5_ICA/ICA mix 2.wav",'r')

#Extracting the frame of the audio and converting it into integer format
signal_2_string=audio2.readframes(-1)
signal_2=np.fromstring(signal_2_string,'Int16')

fr2=audio2.getframerate()
time2=np.linspace(0,len(signal_2)/fr2,len(signal_2))

plt.figure(figsize=(12,2))
plt.plot(time2,signal_2,c="#3ABFE7")
plt.ylim(-35000, 35000)

```

```python
#creating a dataset by zipping all the 3 signals

X=list(zip(signal_1,signal_2))

#importing fastICA module
from sklearn.decomposition import FastICA

#making instantiate of the model
ica=FastICA(n_components=2)

result=ica.fit_transform(X)
```

```python
result.shape
```

```python
result_signal1=result[:,0]
result_signal2=result[:,1]
```

```python
plt.figure(figsize=(12,2))
plt.title('Independent Component #1')
plt.plot(result_signal1, c="#df8efd")
plt.ylim(-0.010, 0.010)
plt.show()


plt.figure(figsize=(12,2))
plt.title('Independent Component #1')
plt.plot(result_signal2, c="#df8efd")
plt.ylim(-0.010, 0.010)
plt.show()
```
