### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('D:/Ai/Data_analyst/House_price_prediction/train.csv')
```

```python
print(dataset.shape)
```

```python
dataset.head()
```

```python
features_with_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]
```

# Handling Missing values

```python
for feature in features_with_na:
    data=dataset.copy()
    
    data[feature]=np.where(data[feature].isnull(),1,0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
    
```

Here, we noticed that all the feature with high Nan value have great sales price.

# Finding all the numerical and categorical features in data

```python
numerical_feature=[feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
categorical_feature=[feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
```

```python
print("Total Numerical Features: {}".format(len(numerical_feature)))
print("Total Categorical Features: {}".format(len(categorical_feature)))
```

```python
dataset[numerical_feature].head()
```

Features like YearBuilt,YearRemodAdd,YrSold and GarageYrBlt are temporal variables. Beacuse Every year new data will be coming so we  cannot process it.

```python
dataset[categorical_feature].head()
```

```python
year_feature=[feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
```

```python
year_feature
```

```python

for feature in year_feature:
    dataset.groupby(feature)['SalePrice'].median().plot()
    
    plt.xlabel(feature)
    plt.ylabel('Sales Price median')
    plt.show()
```

```python
for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrete_feature=[feature for feature in numerical_feature if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
```

```python
dataset[discrete_feature].head(10)
```

```python
for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

# Continuous Feature

```python
continuous_feature=[feature for feature in numerical_feature if feature not in discrete_feature+year_feature+['Id'] ]
```

```python
data[continuous_feature].head()
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

We will use lograthmic transformation

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```


# Outliers Detection

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
    
```
