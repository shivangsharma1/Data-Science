# Advance_house_price_prediction_feature_engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sn

pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('D:/Ai/Data_analyst/House_price_prediction/train.csv')
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(dataset,dataset['SalePrice'],test_size=0.1,random_state=0)
```

Finding Nan values in categorical features

```python
feature_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']

for feature in feature_nan:
    print("{}: {} missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
def replace_missing_values(dataset,feature_nan):
    data=dataset.copy()
    data[feature_nan]=data[feature_nan].fillna('Missing')
    return data

dataset=replace_missing_values(dataset,feature_nan)

dataset[feature_nan].isnull().sum()
```

```python
dataset.head()
```

```python
numerical_missing_feature=[feature for feature in dataset.columns if dataset[feature].dtypes!='O' and dataset[feature].isnull().sum()>1]

for feature in numerical_missing_feature:
    print("{} : {} missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

Replacing missing values with median of that feature value

```python
for feature in numerical_missing_feature:
    # Calculating meadian of that feature
    median_value=dataset[feature].median()
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
    
dataset[numerical_missing_feature].isnull().sum()
    
```

```python
dataset.head(10)
```

Handling Temporal Features

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
dataset.head()
```

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

```python
numerical_feature=['LotFrontage','LotArea','1stFlrSF','SalePrice']
for feature in numerical_feature:
    dataset[feature]=np.log(dataset[feature])
```

# Handling Rare categorical feature

```python
categorical_feature=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
```

```python
categorical_feature
```

```python
for feature in categorical_feature:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
    
```
