# Agglomerative Clustering

```python
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram,ward,single
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#loading points for example, can also load some starting points: X=load_iris().data[:50]
X=load_iris().data
```

```python
model=AgglomerativeClustering(n_clusters=3,linkage='ward')
```

```python
labels=model.fit_predict(X)
print(labels)
```

```python
matrix1=ward(X);
dendrogram(matrix1);
#Here in this graph it is clearly visible there are 3 clusters
```

```python
matrix2=single(X);
dendrogram(matrix2);
#here there is no clear cut between the graphs
```
